```python
import pandas as pd
```

```python
path_vcf = '../../data/res/imputed.ann.oneperline.vcf'
path_vcf_res = '../../data/res/vcf.csv'
skiprows = 15
```

```python
df = pd.read_csv(path_vcf, sep='\t', skiprows=skiprows)
```

```python
total_count = len(df.index)
print(total_count)
df.head(4)
```

```python
df_ = df.iloc[:, 9:]
```

```python
df_.head(4)
```

```python
#remove trailing data (all after :)
def clean(x):
    return x[:x.find(":")]
```

```python
df_ = df_.applymap(clean)
```

```python
#translate df to zeros and ones
def translate(x):
    if x == '0/0':
        return 0
    if x == '1/1':
        return 1
    return -1
```

```python
df_ = df_.applymap(translate)
```

```python
df_.head(3)
```

```python
def missing_data(x):
    x_ = list(x)
    missing_data_value = x_.count(-1) * 100 / len(x_)
    return missing_data_value < 5
```

```python
df_ = df_[df_.apply(missing_data, axis=1)]
```

```python
df_['#CHROM'] = df['#CHROM']
df_['POS'] = df['POS']
df_['INFO'] = df['INFO']
df_['REF'] = df['REF']
df_['ALT'] = df['ALT']
```

```python
new_count = len((df.index))
print(new_count)
print(new_count * 100 / total_count)
```

```python
df_.head(3)
```

```python
df_.to_csv(path_vcf_res, sep='\t', index=None)
path_vcf_res
```
